#Various Activation Functions

```
#sigmoid
def sigmoid(X):
  return 1/(1 + np.exp(-X))

#reverse sigmoid
def r_sigmoid(dA, X):
  return dA * (sigmoid * (1-X))

#relu 
def relu(X): 
  return np.maximum(0,X)
```

# INPUTS

```
  #OUTPUTS 
  o1 = 0.01
  o2 = 0.99
  #INPUTS 
  I_1 = .05
  I_2 = .10
  #BIAS
  B_1 = .35
  B_2 =.60
  #Hidden layer 1
  W_1 = .15    
  W_2 = .20
  W_3 = .25 
  W_4 = .30
  #Output layer 
  W_5 = .40
  W_6 = .45
  W_7 = .50
  W_8 = .55
  #Learning Rate
  L_rate = .50
```

# Foward and Back Propigation Model 

```
#HIDDEN LAYER CALCULATIONS
```

```
n = int(input("How many iteration?"))
for i in range(n):
  
  net_h1 = (I_1 * W_1)+(I_2 * W_2)+(B_1 * 1)
  out_h1 = sigmoid(net_h1)
  net_h2 = (I_1 * W_3)+(I_2 * W_4)+(B_1 * 1)
  out_h2 = sigmoid(net_h2)

  net_o1 = (out_h1 * W_5)+(out_h2 * W_6)+(B_2 * 1)
  out_o1 = sigmoid(net_o1)
  net_o2 = (out_h1 * W_7)+(out_h2 * W_8)+(B_2 * 1)
  out_o2 = sigmoid(net_o2)

  Eo1 = (1/2)*(o1-out_o1)**2
  Eo2 = (1/2)*(o2-out_o2)**2
  E_total = Eo1+Eo2

  Cha_Eo1 = -(o1 - (out_o1))
  Cha_No1 = out_o1*(1-out_o1)
  Cha_NW56 = out_h1

  T56_Cha = (Cha_Eo1 * Cha_No1 * Cha_NW56)
  
  Cha_Eo2 = -(o2 - (out_o2))
  Cha_No2 = out_o2*(1-out_o2)
  Cha_NW78 = out_h2

  T78_Cha = (Cha_Eo2 * Cha_No2 * Cha_NW78)

  Cha_Eh1 = ((Cha_Eo1 * Cha_No1* W_5)+ (Cha_Eo2* Cha_No2 * W_7))
  Cha_Nh1 = out_h1*(1-out_h1)
  Cha_NW12 = I_1

  T12_Cha = (Cha_Eh1 * Cha_Nh1 * Cha_NW12)

  Cha_Eh2 = ((Cha_Eo2 * Cha_No2* W_6)+ (Cha_Eo2* Cha_No2 * W_8))
  Cha_Nh2 = out_h2*(1-out_h2)
  Cha_NW34 = I_2

  T34_Cha = (Cha_Eh2 * Cha_Nh2 * Cha_NW34)

  W_1 = W_1 - L_rate * T12_Cha
  W_2 = W_2 - L_rate * T12_Cha
  W_3 = W_3 - L_rate * T34_Cha
  W_4 = W_4 - L_rate * T34_Cha
  W_5 = W_5 - L_rate * T56_Cha
  W_6 = W_6 - L_rate * T56_Cha
  W_7 = W_7 - L_rate * T78_Cha
  W_8 = W_8 - L_rate * T78_Cha


  print("OUT 1",out_o1)
  print("OUT 2",out_o2)
```
